```python
import matplotlib.pyplot as plt
import re
import numpy as np

ptrn = re.compile(r'[\[\]\/]')
d = []
with open('log.txt') as f:
    for line in f.readlines():
        line = ptrn.sub(' ', line).split()
        _, epoch, _, _, _, tr, _, _, val, _ = line
        d.append([epoch, tr, val])
x, t, v = np.array(d, dtype=float).transpose()
```

```python
f, ax = plt.subplots(1,1, figsize=(12,5))
ax.plot(x, t, 'b', label='train')
ax.plot(x, v, 'r', label='valid')
ax.grid('both')
ax.legend(loc='upper right', fontsize=12)
ax.set_ylim([0.1,0.4])
```
